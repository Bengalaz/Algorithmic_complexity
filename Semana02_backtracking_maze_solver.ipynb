```python
from algorithmic_complexity.aclib.mazebuilder import makeMaze
#graficos de histogramas, barras , boxplots y la mayoria usa ejes x,y
import matplotlib.pyplot as plt
```

Tip : usamos matplotlib para dibujar laberinto , puedes experimentar con [otros colores](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
def drawMaze(maze):
    fig, ax = plt.subplots(figsize=(18,18))
    ax.imshow(maze, cmap='viridis')
    #sin ejes
    ax.axis('off')
```

```python
def mazeSolve(maze, row, col, rowF,colF):
    maze[row][col] = 3       #camino correcto(amarillo)
    if row==rowF and col==colF:
        return True
    
    for r,c in [(row-1,col),(row,col+1),(row+1,col),(row,col-1)]:
        if maze[r][c] ==1 and mazeSolve(maze,r,c,rowF,colF):
            return True
    
    maze[row][col] = 2   #camino sin salida(verde)
    return False
```

```python
maze = makeMaze(5,10)
drawMaze(maze)
```

```python
rows, cols = maze.shape #es una matriz numpy,sino seria len(maze), len(maze[0]) en lugar de maze.shape
mazeSolve(maze, 1, 1,rows-2,cols-2)
drawMaze(maze)
```

```python
maze = makeMaze(60,120)
drawMaze(maze)
```

```python
rows, cols = maze.shape #es una matriz numpy,sino seria len(maze), len(maze[0]) en lugar de maze.shape
mazeSolve(maze, 1, 1,rows-2,cols-2)
drawMaze(maze)
#maximo 1000 llamadas recursivas para este programa
```
